#### 6. Clustering the wasserstein distance matrix

````python
twoclsm = KMeans(n_clusters=2)
aux = dm.diffusion_map.DiffusionMap.from_sklearn(metric='precomputed',
                                                 n_evecs = 2, 
                                                 epsilon =1, 
                                                 alpha = 0.5, 
                                                 k=64)
aux = aux.fit_transform(data.wdist['HTD_0.5'])
````

````python
_,ax1 = plt.subplots(1,2,figsize=(10,4))
ax1 = ax1.ravel()
sns.scatterplot(x=aux[:,0],y=aux[:,1],hue=twoclsm.fit_predict(data.wdist['HTD_0.5']),ax=ax1[0]) 
ax1[0].legend(loc='upper center', bbox_to_anchor=(0.5, -0.15),
          fancybox=True, shadow=True, ncol=5)
ax1[0].set_title('Kmeans in the $W$-space')
ax1[0].set_xlabel('DM1')
ax1[0].set_ylabel('DM2')
sns.scatterplot(x=aux[:,0],y=aux[:,1],hue=pdata_metadata.loc[data.Cs['PCA'].index,'accLabel'],ax=ax1[1])
ax1[1].legend(loc='upper center', bbox_to_anchor=(0.5, -0.15),
          fancybox=True, shadow=True, ncol=5)
ax1[1].set_title('Original Labels')
ax1[1].set_xlabel('DM1')
ax1[1].set_ylabel('DM2')
````

#### 7. Explaining our clusters

````python
pdata_metadata.loc[data.p.columns,'cls'] = twoclsm
````

````python
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_regression
import numpy as np
regr = RandomForestClassifier(max_depth=2, random_state=0)
edist = data.p.T
regr.fit(edist, twoclsm)
sel =data.p.iloc[np.argsort(regr.feature_importances_)[::-1][0:15],]
plt.figure()
sns.clustermap(sel.T,cmap='RdYlBu_r',row_cluster=True)
plt.show()
````

####  8. Features per Samples

````python
fig2,ax2 = plt.subplots(3,5,figsize=(30,12),dpi=300)
ax2 = ax2.ravel()
it=regr.feature_names_in_[np.argsort(regr.feature_importances_)[::-1][0:15]]
for idi, i in enumerate(it):
    sns.scatterplot(x=aux[:,0],
                    y=aux[:,1],
                    hue=edist.T.loc[i,],
                    ax=ax2[idi],legend=False,palette='RdYlBu_r',s=100)
    ax2[idi].set_title(f"{i}",fontdict={"fontsize":20})
````

````python
import numpy.linalg as lg
````

````python
wscores = data.p @ lg.svd(data.wdist['HTD_0.5'])[1].reshape(len(lg.svd(data.wdist['HTD_0.5'])[1]),1)
````

````python
fig2,ax2 = plt.subplots(3,5,figsize=(30,12),dpi=300)
ax2 = ax2.ravel()
it=wscores.sort_values(0,ascending=False).iloc[0:15].index.tolist()
for idi, i in enumerate(it):
    sns.scatterplot(x=aux[:,0],
                    y=aux[:,1],
                    hue=edist.T.loc[i,],
                    ax=ax2[idi],legend=False,palette='RdYlBu_r',s=100)
    ax2[idi].set_title(f"{i}",fontdict={"fontsize":20})
````
