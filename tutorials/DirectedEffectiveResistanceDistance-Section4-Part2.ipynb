```python
import networkx as nx
import scipy.linalg as slg
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn import preprocessing
from sklearn.manifold import MDS
from accordion import distances
import seaborn as sns
```

```python
G = nx.DiGraph()
G.add_edge(1,0,weight=0.5)
G.add_edge(2,0,weight=0.5)
```

```python
len(G.edges())
```

```python
D = distances.ctd_dist(G)
nx.draw_random(G,node_color=D[1,:],cmap=plt.cm.viridis,with_labels=True)
```

```python
sns.heatmap(D) 
```

```python
G = nx.DiGraph()
for i in range(21):
    if i == 20:
        G.add_edge(i,0,weight=0.5)
    else:
        G.add_edge(i,i+1,weight=0.5)
        
for i in range(20,49):
        G.add_edge(i,i+1,weight=0.5)
        G.add_edge(i+1,i,weight=0.5)
        if i == 48:
            G.add_edge(49,20,weight=0.5)
G.add_edge(20,49)
            
D = distances.ctd_dist(G)
nx.draw_kamada_kawai(G,node_color=D[10,:],cmap=plt.cm.viridis)
```

```python
plt.imshow(D,cmap='Reds',interpolation='nearest')
```
